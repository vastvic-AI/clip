```python
import torch
from torchvision import datasets, transforms
import clip
import numpy as np
import os
import torch
from torchvision.datasets import CIFAR100
from sklearn.metrics.pairwise import cosine_similarity

# Define a transform to normalize the data
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

# Load the CIFAR-100 dataset
#trainset = datasets.CIFAR100(root='./data', train=True, download=True, transform=transform)
#trainloader = torch.utils.data.DataLoader(trainset, batch_size=4, shuffle=True, num_workers=2)
cifar100 = CIFAR100(root=os.path.expanduser("~/.cache"), download=True, train=False)

# Load the model and the preprocessing function
device = "cuda" if torch.cuda.is_available() else "cpu"
model, preprocess = clip.load("ViT-B/32", device)

image_features = []
text_prompts = ["cat", "dog", "deer","bird","frog","horse","ship"]  # Add more as needed
text_features = []

# Preprocess the text prompts
text_inputs = clip.tokenize(text_prompts).to(device)

# Process images in the dataset
for image, labels in cifar100:
    # for image in images:
    image = preprocess(image).unsqueeze(0).to(device)
    with torch.no_grad():
        image_feature = model.encode_image(image)
        image_features.append(image_feature.cpu().numpy())

with torch.no_grad():
    text_features = model.encode_text(text_inputs).cpu().numpy()
similarities = cosine_similarity(np.array(image_features).squeeze(), np.array(text_features).squeeze())

for i, image_feature in enumerate(image_features):
    print(f"Image {i} similarities:")
    for j, text in enumerate(text_prompts):
        print(f"   Text '{text}': {similarities[i][j]:.4f}")
```
